```python
# Aplicando a imputa챌찾o de dados para preencher os valores ausentes
dadosRH['educacao'].fillna(dadosRH['educacao'].mode()[0], inplace = True)
dadosRH['aval_ano_anterior'].fillna(dadosRH['aval_ano_anterior'].median(), inplace = True)
```

```python
# Analisando os dados ap처s a imputa챌찾o
dadosRH.isnull().sum()
```

```python
# Verificando o balanceamento de classe na vari찼vel "promovido"
dadosRH.groupby(['promovido']).count()
```

```python
sns.countplot(dadosRH['promovido'])
```

```python
df_classe_majoritaria = dadosRH[dadosRH.promovido==0]
df_classe_minoritaria = dadosRH[dadosRH.promovido==1]
```

```python
df_classe_majoritaria.shape
```

```python
df_classe_minoritaria.shape
```

```python
# Upsample da classe minorit찼ria
from sklearn.utils import resample
df_classe_minoritaria_upsampled = resample(df_classe_minoritaria,
                                          replace= True,
                                          n_samples = 50140,
                                          random_state = 150)
```

```python
dadosRH_balanceados = pd.concat([df_classe_majoritaria, df_classe_minoritaria_upsampled])
```

```python
dadosRH_balanceados.promovido.value_counts()
```

```python
dadosRH_balanceados.info()
```

```python
sns.countplot(dadosRH_balanceados['promovido'])
```

```python
# Uma vez balanceados os dados, vamos salvar o dataset.
dadosRH_balanceados.to_csv('dadosRH_modificado.csv', encoding = 'utf-8', index = False)
```

```python
# Para carregar os dados e continuar o trabalho de an찼lise futuramente
dataset = pd.read_csv('dadosRH_modificado.csv')
dataset.head()
```

# Pergunta 1 - Qual a correla챌찾o entre os atributos dos funcion찼rios?

```python
corr = dataset.corr()
sns.heatmap(corr, cmap = "YlOrRd", linewidths = 0.1)
plt.show()
```

# Pergunta 2 - Qual o tempo de servi챌o da maioria dos funcion찼rios?

```python
sns.distplot(dataset['tempo_servico'], color = "green")
plt.title('Distribui챌찾o do tempo de servi챌o dos funcion찼rios', fontsize = 15)
plt.xlabel('Tempo de servi챌o em anos', fontsize = 15)
plt.ylabel('Total')
plt.show()
```

# Pergunta 3  - Qual avalia챌찾o do ano anterior foi mais comum?

```python
dataset['aval_ano_anterior'].value_counts().sort_values().plot.bar(color ="blue", figsize=(10,5))
plt.title('Distribui챌찾o da avalia챌찾o do ano anterior dos funcion찼rios', fontsize=15)
plt.xlabel('Avalia챌천es', fontsize=15)
plt.ylabel('Total')
plt.show()
```

# Pergunta 4 - Qual a distribui챌찾o das idades dos funcion찼rios?

```python
sns.distplot(dataset['idade'], color = 'magenta')
plt.title('Distribui챌찾o da idade dos funcion찼rios', fontsize=15)
plt.xlabel("Idade", fontsize = 15)
plt.ylabel("Total")
plt.show()
```

# Pergunta 5 - Qual o n첬mero de treinamentos mais frequente?

```python
sns.violinplot(dataset['numero_treinamentos'], color = 'red')
plt.title("N첬mero de treinamentos feitos pelos funcion찼rios", fontsize=15)
plt.ylabel("N첬mero de treinamentos", fontsize=15)
plt.xlabel("Frequ챗ncia")
plt.show()
```

# Pergunta 6 - Qual a propor챌찾o dos funcion찼rios por canal de recrutamento?

```python
dataset['canal_recrutamento'].value_counts()
fatias = [55375, 42358, 2547]
labels = "Outro", "Outsourcing", "Indica챌찾o"
colors = ["purple", "lime", "yellow"]
explode = [0, 0, 0]
plt.pie(fatias, labels = labels, colors = colors, explode = explode, shadow = True, autopct = "%.2f%%")
plt.title("Percentual de funcion찼rios por canal de recrutamento", fontsize=15)
plt.axis("off")
plt.legend()
plt.show()
```

# Pergunta 7 - Qual a rela챌찾o entre a promo챌찾o e a avalia챌찾o no ano anterior?

```python
data = pd.crosstab(dataset['aval_ano_anterior'], dataset['promovido'])
data.div(data.sum(1).astype(float), axis = 0).plot(kind = 'bar',
                                                  stacked = True,
                                                   figsize = (16,9),
                                                   color = ["blue", "magenta"])
plt.title("Rela챌찾o entre avalia챌찾o do ano anterior e a promo챌찾o", fontsize=15)
plt.xlabel("Avalia챌찾o do ano anterior", fontsize=15)
plt.legend()
plt.show()
```
